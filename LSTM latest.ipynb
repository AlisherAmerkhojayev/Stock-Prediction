```python
# Import Libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
import keras_tuner as kt
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_percentage_error
import random

# Set random seed (Original)
seed_value = 50
tf.random.set_seed(seed_value)
np.random.seed(seed_value)
random.seed(seed_value)

#Load and Preprocess the Dataset (Original)
data = pd.read_csv(r"C:\Users\Alisher Amer\OneDrive\Desktop\Project DS\FTSE100.csv")

# Ensure 'Date' column is in datetime format and set it as index (Original)
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)
data.set_index('Date', inplace=True)

# Select features and target (Open price) (Original)
features = ['Close', 'High', 'Low', 'Volume', 'Interest rate', 'Inflation (CPI)',
            'GDP growth rate', 'Unemployment rate', 'Exchange rate to EUR', 'Exchange rate to USD']
target = 'Open'

# Split the data into training (70%) and test (30%)  (Original)
X_lstm = data[features].values
y_lstm = data[target].values
X_train_lstm, X_unseen_lstm = X_lstm[:-830], X_lstm[-830:]
y_train_lstm, y_unseen_lstm = y_lstm[:-830], y_lstm[-830:]

# Initialize separate MinMaxScalers for X and y (Taken from Sklearn documentation)
scaler_X_lstm = MinMaxScaler()
scaler_y_lstm = MinMaxScaler()

#Fit the scaler for X ONLY on the training data and transform it (Taken: MinMaxScaler)
X_train_lstm_scaled = scaler_X_lstm.fit_transform(X_train_lstm)

#Fit the scaler for y ONLY on the training data and transform it (Taken: MinMaxScaler)
y_train_lstm_scaled = scaler_y_lstm.fit_transform(y_train_lstm.reshape(-1, 1))

#Transform the test data using the already fitted scaler for X and y (Taken: MinMaxScaler)
X_unseen_lstm_scaled = scaler_X_lstm.transform(X_unseen_lstm)
y_unseen_lstm_scaled = scaler_y_lstm.transform(y_unseen_lstm.reshape(-1, 1))

# Create a sliding window function for time series data (Original)
def create_sliding_window(data, target, window_size):
    X, y = [], []
    for i in range(window_size, len(data)):
        X.append(data[i - window_size:i])
        y.append(target[i])
    return np.array(X), np.array(y)

#Apply sliding window to scaled training data (X_train_windowed, y_train_windowed, X_unseen_windowed, y_unseen_windowed) (Original)
window_size = 20
X_train_lstm_windowed, y_train_lstm_windowed = create_sliding_window(X_train_lstm_scaled, y_train_lstm_scaled, window_size)
X_unseen_lstm_windowed, y_unseen_lstm_windowed = create_sliding_window(X_unseen_lstm_scaled, y_unseen_lstm_scaled, window_size)

# Build CNN model using Keras Tuner for hyperparameter tuning (Taken from Keras Tuner Documentation and Keras CNN API)
def build_lstm_model(hp):
    model = Sequential()
    # First LSTM layer
    model.add(LSTM(
        units=hp.Int('units1', min_value=32, max_value=256, step=32),
        activation='tanh',
        return_sequences=hp.Choice('return_seq1', [True, False]),
        input_shape=(X_train_lstm_windowed.shape[1], X_train_lstm_windowed.shape[2])
    ))
    model.add(Dropout(hp.Float('dropout1', min_value=0.3, max_value=0.5, step=0.1)))

    # Second LSTM layer (optional)
    if hp.Boolean('use_second_lstm'):
        model.add(LSTM(
            units=hp.Int('units2', min_value=32, max_value=256, step=32),
            activation='tanh',
            return_sequences=False
        ))
        model.add(Dropout(hp.Float('dropout2', min_value=0.3, max_value=0.5, step=0.1)))
    else:
        if hp.get('return_seq1'):
            model.add(tf.keras.layers.Flatten())

    # Dense layer
    model.add(Dense(units=hp.Int('dense_units', min_value=16, max_value=128, step=16), activation='relu'))
    model.add(Dropout(hp.Float('dropout_dense', min_value=0.3, max_value=0.5, step=0.1)))

    # Output layer
    model.add(Dense(1))

    # Compile the model
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])),
                  loss='mean_squared_error')
    return model

# Define additional evaluation metrics (Original)

def smape(actual, predicted):

    return 100 / len(actual) * np.sum(2 * np.abs(predicted - actual) / (np.abs(actual) + np.abs(predicted)))


def mean_bias_deviation(actual, predicted):

    return np.mean(predicted - actual)


def directional_accuracy(actual, predicted):

    return np.mean(np.sign(actual[1:] - actual[:-1]) == np.sign(predicted[1:] - predicted[:-1])) * 100


def accuracy_within_tolerance(y_true, y_pred, tolerance=0.05):

    correct = np.sum(np.abs(y_true - y_pred) <= tolerance * y_true)
    return correct / len(y_true) * 100


def exact_accuracy(y_true, y_pred, threshold=0.1):

    correct = np.sum(np.abs(y_true - y_pred) <= threshold)
    return correct / len(y_true) * 100

# 3. Hyperparameter tuning with Keras Tuner (Taken from  Keras Tuner Documentation)
tuner = kt.RandomSearch(
    build_lstm_model,
    objective='val_loss',
    max_trials=30,
    executions_per_trial=1,
    directory='tuner_dir_lstm_00',
    project_name='lstm_stock_prediction_00'
)

# Initialize lists to store metrics over multiple iterations for training/validation data (Original)
mae_list_train, mse_list_train, rmse_list_train, r2_list_train = [], [], [], []
mape_list_train, smape_list_train, explained_variance_list_train = [], [], []
mbd_list_train, directional_acc_list_train, accuracy_within_1_list_train, exact_acc_list_train = [], [], [], []

# Initialize lists to store metrics over multiple iterations for test (unseen) data (Original)
mae_list_test, mse_list_test, rmse_list_test, r2_list_test = [], [], [], []
mape_list_test, smape_list_test, explained_variance_list_test = [], [], []
mbd_list_test, directional_acc_list_test, accuracy_within_1_list_test, exact_acc_list_test = [], [], [], []
unseen_predictions_list = []

# Number of iterations (Original)
n_iterations = 100  

# Function to calculate metrics and append results to lists (Original)
def evaluate_model(actual, predicted, dataset_type="Training_Validation"):
    actual_prices = scaler_y_lstm.inverse_transform(actual.reshape(-1, 1))
    predicted_prices = scaler_y_lstm.inverse_transform(predicted.reshape(-1, 1))

    # Calculate evaluation metrics (Taken from Sklearn Metrics)
    mae = mean_absolute_error(actual_prices, predicted_prices)
    mse = mean_squared_error(actual_prices, predicted_prices)
    rmse = np.sqrt(mse)
    r2 = r2_score(actual_prices, predicted_prices)
    mape = mean_absolute_percentage_error(actual_prices, predicted_prices)
    smape_value = smape(actual_prices, predicted_prices)
    explained_variance = explained_variance_score(actual_prices, predicted_prices)
    mbd = mean_bias_deviation(actual_prices, predicted_prices)
    directional_acc = directional_accuracy(actual_prices, predicted_prices)
    accuracy_within_1 = accuracy_within_tolerance(actual_prices, predicted_prices, tolerance=0.01)
    exact_acc = exact_accuracy(actual_prices, predicted_prices, threshold=50)

    # Append metrics to respective lists based on the dataset_type (Original)
    if dataset_type == "Training_Validation":
        mae_list_train.append(mae)
        mse_list_train.append(mse)
        rmse_list_train.append(rmse)
        r2_list_train.append(r2)
        mape_list_train.append(mape)
        smape_list_train.append(smape_value)
        explained_variance_list_train.append(explained_variance)
        mbd_list_train.append(mbd)
        directional_acc_list_train.append(directional_acc)
        accuracy_within_1_list_train.append(accuracy_within_1)
        exact_acc_list_train.append(exact_acc)
    elif dataset_type == "Unseen_LSTM":
        mae_list_test.append(mae)
        mse_list_test.append(mse)
        rmse_list_test.append(rmse)
        r2_list_test.append(r2)
        mape_list_test.append(mape)
        smape_list_test.append(smape_value)
        explained_variance_list_test.append(explained_variance)
        mbd_list_test.append(mbd)
        directional_acc_list_test.append(directional_acc)
        accuracy_within_1_list_test.append(accuracy_within_1)
        exact_acc_list_test.append(exact_acc)

    # Print out the evaluation metrics for each iteration (Original)
    print(f"\n{dataset_type} Data Evaluation Metrics:")
    print(f"Mean Absolute Error (MAE): {mae}")
    print(f"Mean Squared Error (MSE): {mse}")
    print(f"Root Mean Squared Error (RMSE): {rmse}")
    print(f"R-squared (R²): {r2}")
    print(f"Mean Absolute Percentage Error (MAPE): {mape}")
    print(f"Symmetric Mean Absolute Percentage Error (SMAPE): {smape_value}")
    print(f"Explained Variance Score: {explained_variance}")
    print(f"Mean Bias Deviation (MBD): {mbd}")
    print(f"Directional Accuracy: {directional_acc}%")
    print(f"Accuracy within 1% tolerance: {accuracy_within_1}%")
    print(f"Exact Accuracy (within 50 units): {exact_acc}%")

# Perform multiple iterations of model training and evaluation (Original)
for iteration in range(n_iterations):
    print(f"Iteration {iteration + 1}/{n_iterations}")

    ensemble_predictions_list = []
    actuals_list = []

    # TimeSeriesSplit for Cross-Validation on Training Data (Original)
    tscv = TimeSeriesSplit(n_splits=5)

    # Train the model using cross-validation
    for fold, (train_index, test_index) in enumerate(tscv.split(X_train_lstm_windowed)):
        print(f"Training fold {fold + 1}...")
        X_cv_train, X_cv_test = X_train_lstm_windowed[train_index], X_train_lstm_windowed[test_index]
        y_cv_train, y_cv_test = y_train_lstm_windowed[train_index], y_train_lstm_windowed[test_index]

        # Early Stopping Taken from Keras EarlyStopping API)
        early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

        # Perform the hyperparameter search (Taken from Keras Tuner documentation)
        tuner.search(X_cv_train, y_cv_train, epochs=50, batch_size=32, validation_data=(X_cv_test, y_cv_test),
                     callbacks=[early_stopping], verbose=0)

        # Get the best hyperparameters (Taken from Keras Tuner documentation)
        best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

        # Build the best model using the best hyperparameters (Taken from Keras Tuner documentation)
        best_model = tuner.hypermodel.build(best_hps)

        # Train the best model (Original)
        best_model.fit(X_cv_train, y_cv_train, epochs=100, batch_size=32, validation_data=(X_cv_test, y_cv_test),
                       callbacks=[early_stopping], verbose=0)

        # Predict the test set (Original)
        predictions = best_model.predict(X_cv_test)

        # Store the predictions and actuals (Original)
        ensemble_predictions_list.append(predictions)
        actuals_list.append(y_cv_test)

    # Combine predictions and actual values for training/validation (Original)
    ensemble_predictions = np.concatenate(ensemble_predictions_list, axis=0)
    actuals = np.concatenate(actuals_list, axis=0)

    # Evaluate the model on the training/validation set for this iteration (Original)
    evaluate_model(actuals, ensemble_predictions, dataset_type="Training_Validation")
    
    # Evaluate on test data (Original)
    unseen_predictions_lstm = best_model.predict(X_unseen_lstm_windowed)
    evaluate_model(y_unseen_lstm_windowed, unseen_predictions_lstm, dataset_type="Unseen_LSTM")

# After the iterations, compute and print summary statistics (Original)
def print_summary_statistics(metric_list, metric_name, dataset_type):
    print(f"\nSummary Statistics for {metric_name} ({dataset_type}):")
    print(f"Mean: {np.mean(metric_list):.4f}")
    print(f"Median: {np.median(metric_list):.4f}")
    print(f"Standard Deviation: {np.std(metric_list):.4f}")

# Print summary statistics for training/validation data (Original)
print("\nTraining/Validation Data Summary Statistics:")
print_summary_statistics(mae_list_train, "Mean Absolute Error (MAE)", "Training/Validation")
print_summary_statistics(mse_list_train, "Mean Squared Error (MSE)", "Training/Validation")
print_summary_statistics(rmse_list_train, "Root Mean Squared Error (RMSE)", "Training/Validation")
print_summary_statistics(r2_list_train, "R-squared (R²)", "Training/Validation")
print_summary_statistics(mape_list_train, "Mean Absolute Percentage Error (MAPE)", "Training/Validation")
print_summary_statistics(smape_list_train, "Symmetric Mean Absolute Percentage Error (SMAPE)", "Training/Validation")
print_summary_statistics(explained_variance_list_train, "Explained Variance Score", "Training/Validation")
print_summary_statistics(mbd_list_train, "Mean Bias Deviation (MBD)", "Training/Validation")
print_summary_statistics(directional_acc_list_train, "Directional Accuracy", "Training/Validation")
print_summary_statistics(accuracy_within_1_list_train, "Accuracy within 1% tolerance", "Training/Validation")
print_summary_statistics(exact_acc_list_train, "Exact Accuracy (within 50 units)", "Training/Validation")

# Print summary statistics for unseen test data (Original)
print("\nUnseen Test Data Summary Statistics:")
print_summary_statistics(mae_list_test, "Mean Absolute Error (MAE)", "Unseen Test")
print_summary_statistics(mse_list_test, "Mean Squared Error (MSE)", "Unseen Test")
print_summary_statistics(rmse_list_test, "Root Mean Squared Error (RMSE)", "Unseen Test")
print_summary_statistics(r2_list_test, "R-squared (R²)", "Unseen Test")
print_summary_statistics(mape_list_test, "Mean Absolute Percentage Error (MAPE)", "Unseen Test")
print_summary_statistics(smape_list_test, "Symmetric Mean Absolute Percentage Error (SMAPE)", "Unseen Test")
print_summary_statistics(explained_variance_list_test, "Explained Variance Score", "Unseen Test")
print_summary_statistics(mbd_list_test, "Mean Bias Deviation (MBD)", "Unseen Test")
print_summary_statistics(directional_acc_list_test, "Directional Accuracy", "Unseen Test")
print_summary_statistics(accuracy_within_1_list_test, "Accuracy within 1% tolerance", "Unseen Test")
print_summary_statistics(exact_acc_list_test, "Exact Accuracy (within 50 units)", "Unseen Test")

# Plot the directional accuracy and RMSE over the iterations for both Training/Validation and Unseen data (Original)
plt.figure(figsize=(12, 6))

# Plot Directional Accuracy over iterations for Training/Validation data (Original)
plt.subplot(1, 2, 1)
plt.plot(range(1, n_iterations + 1), directional_acc_list_train, marker='o', color='green', label='Train/Validation')
plt.plot(range(1, n_iterations + 1), directional_acc_list_test, marker='o', color='orange', label='Unseen')
plt.title('Directional Accuracy over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Directional Accuracy (%)')
plt.legend()

# Plot RMSE over iterations for Training/Validation data (Original)
plt.subplot(1, 2, 2)
plt.plot(range(1, n_iterations + 1), rmse_list_train, marker='o', color='blue', label='Train/Validation')
plt.plot(range(1, n_iterations + 1), r2_list_test, marker='o', color='red', label='Unseen')
plt.title('RMSE over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Root Mean Squared Error (RMSE)')
plt.legend()

plt.tight_layout()
plt.show()
# Combine and average unseen predictions across iterations (Original)
combined_unseen_predictions_avg = np.mean(unseen_predictions_list, axis=0)

# Inverse transform the average unseen predictions and actuals back to the original price scale (Original)
combined_unseen_predictions_avg_prices = scaler_y_lstm.inverse_transform(combined_unseen_predictions_avg.reshape(-1, 1))
actual_unseen_prices = scaler_y_lstm.inverse_transform(y_unseen_lstm_windowed.reshape(-1, 1))

# Plot the average unseen predictions vs actual prices (Original)
plt.figure(figsize=(10, 6))
plt.plot(combined_actuals_unseen_avg_prices, label='Actual Unseen_LSTM Prices', color='blue')
plt.plot(combined_unseen_predictions_avg_prices, label='Predicted Unseen_LSTM Prices (Avg)', color='red')
plt.title('Average Predicted vs Actual FTSE 100 Stock Prices (Unseen_LSTM)')
plt.xlabel('Days')
plt.ylabel('Price')
plt.legend()
plt.show()
```
